### Geometry Managers - pack, place, grid

```python
#Button Example - Grade calculation - Uses pack
root = Tk()
root.title="Button using pack"#Configure any widget after creating like this

#Create Labels
lb1 =Label(root,text="Mark1")
lb2 =Label(root,text="Mark2")
lb3 =Label(root,text="Mark3")

#Create text boxes
e1 =Entry(root)
e2 =Entry(root)
e3 =Entry(root)

#Define event handler - function for grade calculation
def grade():
    m1 = int(e1.get())
    m2 = int(e2.get())
    m3 = int(e3.get())
    avg = (m1+m2+m3)/3
    if avg > 90:
        grd = "A"
    elif avg >=80 and avg <= 89:
        grd = "C"
    elif avg >=70 and avg <= 79:
        grd = "C"
    elif avg >=60 and avg <= 69:
        grd = "D"
    elif avg >=50 and avg <= 59:
        grd = "E"
    else:
        grd = "Fail"
    messagebox.showinfo("Your grade is",grd)

#Create a button
B1 = Button(root, text="display", command=grade)

#Arrange on screen
#pack on LEFT,RIGHT,BOTTOM,TOP(default)
#pack first clip last
lb1.pack(side=LEFT)
e1.pack(side=RIGHT)

lb2.pack(side=LEFT)
e2.pack(side=RIGHT)

lb3.pack(side=LEFT)
e3.pack(side=RIGHT)
B1.pack(side=BOTTOM)
root.mainloop()
```

### Control  - Menubutton
Used to display menus

```python
from tkinter import *
from tkinter import messagebox

root= Tk()#Create root
root.title('Menu Demo')#Optional
root.geometry("200x200")#Optional

#Event handlers
def new():
    print("New Menu!")
def disp():
    print("Open Menu!")
    
menubutton = Menubutton(root, text="File")#Create a menu button
m = Menu(menubutton, tearoff = 0)#Create a menu
menubutton.menu = m #Set menubutton's menu property - 
menubutton["menu"] = menubutton.menu #Set menubutton's menu property

#Add commands to the menubutton's menu object
menubutton.menu.add_command(label="Create new",command=new)
menubutton.menu.add_command(label="Open",command=disp)
menubutton.menu.add_separator()#Insert a line in between
menubutton.menu.add_command(label="Exit",command=root.quit)
menubutton.pack()
mainloop()
```

### Control  - Checkbutton
Used to select multiple options from a group of options

```python
### Control 2 - CheckButton
from tkinter import * 
from tkinter import messagebox
root= Tk()
root.title('Checkbutton Demo') 

#Create integer variables to bind to a control
v1=IntVar()#tkinter int variable
v2=IntVar()
v3=IntVar()

#Event handlers
def music():
    if(v1.get()==1):#Select or deselect
        if v2.get()==1:
            cb2.toggle()#Deselect dance of selected already
        print("Music")
def dance():
    if(v2.get()==1):
        if v1.get()==1:
            cb1.toggle()#Deselect music if selected already
        print("Dance")   
def instrument():
    if(v3.get()==1):
        print("Instrument")
def summary():
    lst=[]
    if v1.get()==1:
        lst.append(cb1.cget("text"))
    if v2.get()==1:
        lst.append(cb2.cget("text"))
    if v3.get()==1:
        lst.append(cb3.cget("text"))
    messagebox.showinfo("Selection",",".join(lst))
cb1=Checkbutton(root,text='Music', variable=v1,onvalue=1, offvalue=0, command=music)
cb1.grid(row=1)
cb2=Checkbutton(root,text='Dance', variable=v2,onvalue=1, offvalue=0, command=dance)
cb2.grid(row=2)
cb3=Checkbutton(root,text='Instrument', variable=v3,onvalue=1, offvalue=0, command=instrument)
cb3.grid(row=3)
btn1 = Button(root, text = "Summarize",command=summary)
btn1.grid(row=4)
root.mainloop()
```

### Control  - Radiobutton
Used to select one option from a group

```python
root= Tk()
root.geometry("200x200")

#Create variable - one variable per group
radio=IntVar()

#Event handlers
def choice():
    if(radio.get()==1):
        root.configure(background='red')
    elif(radio.get()==2):
        root.configure(background='blue')
    elif(radio.get()==3):
        root.configure(background='green')

rb1=Radiobutton(root,text='Red', variable=radio,value=1, command=choice)
rb1.grid(row=0)
rb2=Radiobutton(root,text='Blue', variable=radio,value=2, command=choice)
rb2.grid(row=1)
rb3=Radiobutton(root,text='Green', variable=radio,value=3, command=choice)
rb3.grid(row=3)
root.mainloop() 
```
